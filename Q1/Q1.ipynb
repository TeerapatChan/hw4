### PageRank Class

```python

class PageRank:
    def __init__(self, edge_file):

        self.node_degree = {}
        self.max_node_id = 0
        self.edge_file = edge_file

    def read_edge_file(self, edge_file):
        with open(edge_file) as f:
            for line in f:
                val = line.split('\t')
                yield int(val[0]), int(val[1])

    """
    Step1: Calculate the out-degree of each node and maximum node_id of the graph.
    Store the out-degree in class variable "node_degree" and maximum node id to "max_node_id".
    """
    def calculate_node_degree(self):
        for source, target in self.read_edge_file(self.edge_file):

        ### STEP 1
        ### Implement your code here
        #############################################
            
            # 1a. Update the max_node_id, checking BOTH source and target
            #     This ensures nodes that are only targets are counted.
            self.max_node_id = max(self.max_node_id, source, target)
            
            # 1b. Increment the out-degree for the source node.
            #     .get(source, 0) retrieves the current count or 0 if it's a new node.
            self.node_degree[source] = self.node_degree.get(source, 0) + 1
            
        #############################################

        print("Max node id: {}".format(self.max_node_id))
        print("Total source number: {}".format(len(self.node_degree)))

    def get_max_node_id(self):
        return self.max_node_id

    def run_pagerank(self, node_weights,  damping_factor=0.85, iterations=10):

        # FIX: Initialize pr_values with the provided node_weights (P_0 = Pd)
        # The skeleton's original line was incorrect.
        pr_values = list(node_weights)
        N = self.max_node_id + 1
        
        start_time = time.time()
        
        """ 
        Step2: Implement pagerank algorithm as described in lecture slides and the question.
        ...
        """
        for it in range(iterations):
            
            # --- START OF REQUIRED LOGIC ---
            # 1. Calculate total dangling mass (S_t) from the *previous* iteration
            dangling_mass = 0.0
            for j in range(N):
                # A node 'j' is dangling if it has no outgoing edges,
                # i.e., it's not a key in our node_degree dict.
                if j not in self.node_degree:
                    dangling_mass += pr_values[j]
                    
            # 2. Initialize new_pr_values with Teleport + Dangling contribution
            # This is ( (1-d) + d * S_t ) * Pd
            teleport_base = (1.0 - damping_factor)
            dangling_contribution_base = damping_factor * dangling_mass
            
            new_pr_values = [0.0] * N
            for j in range(N):
                # node_weights[j] is the personalization vector Pd[j]
                new_pr_values[j] = (teleport_base * node_weights[j]) + \
                                   (dangling_contribution_base * node_weights[j])
            # --- END OF REQUIRED LOGIC (PART 1) ---

            # The skeleton's loop handles the "Link Contribution"
            for source, target in self.read_edge_file(self.edge_file):

            ### STEP 2
            ### Implement your code here
            #############################################
                # 3. Add Link Contribution (d * M * P_t)
                #    We only "push" from non-dangling nodes
                if source in self.node_degree:
                    out_degree = self.node_degree[source]
                    contribution = damping_factor * (pr_values[source] / out_degree)
                    new_pr_values[target] += contribution
            #############################################

            # --- START OF REQUIRED LOGIC (PART 2) ---
            # 4. Update pr_values for the next iteration
            #    This was missing from the skeleton.
            pr_values = new_pr_values
            # --- END OF REQUIRED LOGIC (PART 2) ---

            print ("Completed {0}/{1} iterations. {2} seconds elapsed.".format(it + 1, iterations, time.time() - start_time))

        # This return (from the skeleton) is now correct, 
        # as pr_values holds the final updated list.
        return pr_values
```

### Simplified Pagerank

```python
# Options
file = 'network.tsv'                # Input file of the edge list - DO NOT EDIT
command = 'simplified_pagerank'     # Command to run - DO NOT EDIT
damping_factor = 0.85               # Damping factor - submit results for damping_factor = 0.85
iterations = [10,25]                # Number of iterations - sumbit results for iterations = [10,25] 

# Run Simplified PageRank
# DO NOT EDIT
for i in iterations:
    pr = PageRank(file)
    pr.calculate_node_degree()
    max_node_id = pr.get_max_node_id()
    node_weights = np.ones(max_node_id + 1) / (max_node_id + 1)
    result = pr.run_pagerank(node_weights=node_weights, iterations=i, damping_factor=damping_factor)
    dump_results(command, i, result )
```

### Personalized Pagerank

```python
# Options
file = 'network.tsv'                # Input file of the edge list - DO NOT EDIT
command = 'personalized_pagerank'   # Command to run - DO NOT EDIT
damping_factor = 0.85               # Damping factor - submit results for damping_factor = 0.85
iterations = [10,25]                # Number of iterations - sumbit results for iterations = [10,25] 

# Run Personalized PageRank
# DO NOT EDIT
for i in iterations:
    pr = PageRank(file)
    pr.calculate_node_degree()
    max_node_id = pr.get_max_node_id()
    np.random.seed(gtid())
    node_weights = np.random.rand(max_node_id + 1)
    node_weights = node_weights/node_weights.sum()
    result = pr.run_pagerank(node_weights=node_weights, iterations=i, damping_factor=damping_factor)
    dump_results(command, i, result)
```
